```python
%%sql



-- Create the 'hotel' table with NULL support for agent and company columns
CREATE TABLE hotel (
    hotel TEXT,
    is_canceled INTEGER,
    lead_time INTEGER,
    arrival_date_year INTEGER,
    arrival_date_month TEXT,
    arrival_date_week_number INTEGER,
    arrival_date_day_of_month INTEGER,
    stays_in_weekend_nights INTEGER,
    stays_in_week_nights INTEGER,
    adults INTEGER,
    children INTEGER,
    babies INTEGER,
    meal TEXT,
    country TEXT,
    market_segment TEXT,
    distribution_channel TEXT,
    is_repeated_guest INTEGER,
    previous_cancellations INTEGER,
    previous_bookings_not_canceled INTEGER,
    reserved_room_type TEXT,
    assigned_room_type TEXT,
    booking_changes INTEGER,
    deposit_type TEXT,
    agent REAL, -- Allowing for NULL values
    company REAL, -- Allowing for NULL values
    days_in_waiting_list INTEGER,
    customer_type TEXT,
    adr REAL,
    required_car_parking_spaces INTEGER,
    total_of_special_requests INTEGER,
    reservation_status TEXT,
    reservation_status_date TEXT
);
```

```python
 ## Calculate total bookings
%sql SELECT COUNT(*) AS total_bookings FROM hotel;
```

```python
## Basic Statistics
%sql SELECT COUNT(*) AS total_bookings, AVG(lead_time) AS average_lead_time FROM hotel;
```

```python
## Cancellation Analysis
%sql SELECT is_canceled, COUNT(*) AS count_cancellations FROM hotel GROUP BY is_canceled;
```

```python
import sqlite3

# Define the SQL query
query = '''
    SELECT
        arrival_date_year,
        arrival_date_month,
        COUNT(*) AS bookings_per_month,
        100 * COUNT(*) / (SELECT COUNT(*) FROM hotel) AS percentage_of_total
    FROM
        hotel
    GROUP BY
        arrival_date_year, arrival_date_month
    ORDER BY
        arrival_date_year, arrival_date_month;
'''

# Connect to the SQLite database and execute the query
conn = sqlite3.connect('hotel_analysis.db')
result = conn.execute(query).fetchall()

# Display the result
result
```
